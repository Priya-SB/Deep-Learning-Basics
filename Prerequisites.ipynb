Data Encoding Techniques

```python
# using the replace function
def replace_marital(val):
  if val=='single':
    return 0
  else:
    return 1

bank['marital']=bank['marital'].apply(replace_marital,1)
```

```python
bank.head()
```

```python
# using the map function
bank['loan']=bank['loan'].map({'no':0,'yes':1}.get)
```

```python
bank.head()
```

```python
# using the replace function with dictionary
bank=bank.replace({'default':{'no':0,'yes':1}})
```

```python
bank.head()
```

```python
bank['job'].unique()
```

```python
# label encoding using category codes
bank.job = bank.job.astype('category')
bank.job = bank.job.cat.codes
```
