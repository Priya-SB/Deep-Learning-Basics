```python
# Graphical Visualizations

# Line Graph
covid.plot(x="Date", y=["dailydeceased", "totalcritical","ventilatorpatients"])
plt.xticks(rotation = 45)
plt.title("Daily Critical patients and Mortality rate")
plt.show()
```

```python
# Scatter Plot
x = covid["dailysamples"]
y = covid["dailyconfirmed"]

plt.title("Number of COVID positive samples")
plt.scatter(x, y, marker='.', color='r')
plt.show()
```

```python
# Bar Graph
x = covid["Date"].iloc[:20]
y = covid["flu"].iloc[:20]
  
plt.title("Flu patients")
plt.xticks(rotation = 45)
plt.bar(x, y, color='m')
plt.show()
```

```python
# Pie Chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("Latest Ratio")

d=covid["totaldeceased"].iloc[-1]/covid["totalsamples"].iloc[-1]*100
r=covid["totalrecovered"].iloc[-1]/covid["totalsamples"].iloc[-1]*100

langs = ['Recovered', 'Deceased']
students = [r,d]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

```python
# Box Plot
fig1, ax1 = plt.subplots()
ax1.set_title('Daily Recovered')
ax1.boxplot(covid["dailyrecovered"])

fig2, ax2 = plt.subplots()
ax2.set_title('Daily Deaths')
ax2.boxplot(covid["dailydeceased"])
```
